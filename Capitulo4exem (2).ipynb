ESTRUTURANDO DADOS COM NUMPY

```python
import numpy as np

#CRIANDO UM NUMPY ARRAY (NDARRAY)  (nao aceita elementos heterogeneos)
#CRIANDO UM NUMPY ARRAY 1-D
print(np.array([1,2,3]))

#CRIANDO UM NUMPY ARRAY 2-D
print(np.array([[1,2,3],[4,5,6]]))
```

```python
#CRIANDO UMA MATRIZ NUMPY ARRAY
mtz = np.arange(1,101).reshape(10,10)
print(mtz)
```

```python
print(mtz.max()) #maximo
print(mtz.min()) #minimo
print(mtz.mean())
print(mtz.size) #tamanho
print(mtz.shape) #shape
```

```python
#OPERAÇÕES COM UM NUMPY ARRAY
arr1 = np.array([10,20,15]) #gastos com agua
arr2 = np.array([20,25,40]) #gastos com energia

#gastos total de cada mes
print(arr1+arr2)
#gastos total de tudo
print((arr1+arr2).sum())
```

```python
# TRABALHANDO COM NUMEROS ALEATORIOS
# RANDOM
np.random.seed(10) #semente aleatoria - permite gerar valores aleatorios iguais
mtz = np.random.randint(1,11,12).reshape(4,3)   #quero numeros de 1 - 10 gerando 12 numeros aleatorios  // reshape transforma em matriz
print(mtz)

#EXTRAINDO VALORES DAS COLUNAS DA MATRIZ
print(mtz.sum(axis=0))

#EXTRAINDO VALORES DAS LINHAS DA MATRIZ
print(mtz.sum(axis=1))

#EXTRAINDO O VALOR DE GASTOS COM NET
print(mtz.sum(axis=1)[2])
```

```python
#EXTRAIR ELEMENTOS UNICOS
print(np.unique(mtz))
#agora faz contagem de quantos elementos tem naquela matriz
print(np.unique(mtz,return_counts=True))
```
